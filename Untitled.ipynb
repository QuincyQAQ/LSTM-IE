```python
# 必要库导入
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd  # 新增：用于保存CSV

# ==== 1. 二维离散混沌映射生成 ====
a = b = 15
N = 1000
x = [0.1]
y = [0.1]

for _ in range(N - 1):
    x_next = a * np.sin(3 ** (y[-1] ** 2) - x[-1] ** 3)
    y_next = b * np.cos(3 ** (x[-1] ** 2) - y[-1] ** 3)
    x.append(x_next)
    y.append(y_next)

x_series = np.array(x)
y_series = np.array(y)

# ==== 2. 构造LSTM训练数据 ====
seq_len = 10
train_len = 500
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def prepare_data(series):
    X, Y = [], []
    for i in range(train_len - seq_len):
        X.append(series[i:i+seq_len])
        Y.append(series[i+seq_len])
    X = torch.tensor(X, dtype=torch.float32).unsqueeze(-1).to(device)
    Y = torch.tensor(Y, dtype=torch.float32).unsqueeze(-1).to(device)
    return X, Y

Xx, Yx = prepare_data(x_series)
Xy, Yy = prepare_data(y_series)

# ==== 3. 定义LSTM模型 ====
class LSTMForecast(nn.Module):
    def __init__(self, hidden_size=64):
        super().__init__()
        self.lstm = nn.LSTM(input_size=1, hidden_size=hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])

# ==== 4. 模型训练函数（修改：添加学习率余弦下降和loss记录）====
def train_model(X_train, Y_train, model_name, hidden_size=64, epochs=500, lr=1e-2):
    model = LSTMForecast(hidden_size).to(device)
    optimizer = optim.Adam(model.parameters(), lr=lr)
    criterion = nn.MSELoss()
    
    # 新增：创建学习率调度器 - 余弦退火
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-5)
    
    # 新增：创建loss记录列表
    loss_history = []
    lr_history = []  # 新增：记录学习率变化
    
    for ep in range(epochs):
        model.train()
        optimizer.zero_grad()
        output = model(X_train)
        loss = criterion(output, Y_train)
        loss.backward()
        optimizer.step()
        
        # 新增：更新学习率
        scheduler.step()
        
        # 新增：记录每个epoch的loss和学习率
        loss_history.append(loss.item())
        current_lr = optimizer.param_groups[0]['lr']
        lr_history.append(current_lr)
        
        # 每50个epoch打印一次进度（新增学习率显示）
        if (ep+1) % 50 == 0:
            print(f'{model_name} - Epoch [{ep+1}/{epochs}], Loss: {loss.item():.6f}, LR: {current_lr:.2e}')
    
    # 新增：将loss和学习率保存到CSV
    df = pd.DataFrame({
        'epoch': range(1, epochs+1),
        'loss': loss_history,
        'learning_rate': lr_history
    })
    df.to_csv(f'{model_name}_training_history.csv', index=False)
    
    return model


# 修改：添加模型名称参数
model_x = train_model(Xx, Yx, 'model_x')
model_y = train_model(Xy, Yy, 'model_y')

# ==== 5. 使用训练好的模型预测后续序列 ====
def predict_future(model, series, pred_len):
    model.eval()
    window = list(series[train_len - seq_len:train_len])
    preds = []
    for _ in range(pred_len):
        inp = torch.tensor(window, dtype=torch.float32).unsqueeze(0).unsqueeze(-1).to(device)
        with torch.no_grad():
            pred = model(inp).item()
        preds.append(pred)
        window.pop(0)
        window.append(pred)
    return np.concatenate([series[:train_len], preds])

pred_len = 500
x_new = predict_future(model_x, x_series, pred_len)
y_new = predict_future(model_y, y_series, pred_len)

# ==== 6. 绘图，完全复现Fig.4风格 ====
series_pairs = [('x', x_series, x_new),
                ('y', y_series, y_new)]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
labels = ['(a)', '(b)']

for idx, (var, orig, pred) in enumerate(series_pairs):
    ax = axes[idx]
    ax.plot(orig, color='blue', label=f'${var}_n$')
    ax.plot(pred, color='red', linestyle='--', label=f'${var}_n\prime$')
    ax.axvline(train_len, color='gray', linestyle=':', linewidth=1)
    ax.set_xlim(0, len(orig))
    ax.set_xlabel('$n$')
    ax.set_ylabel(f'${var}$')
    ax.legend(loc='upper right')
    ax.text(0.05, 0.9, labels[idx], transform=ax.transAxes, fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# 必要库导入
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd  # 新增：用于保存CSV

# ==== 1. 二维离散混沌映射生成 ====
a = b = 15
N = 1000
x = [0.1]
y = [0.1]

for _ in range(N - 1):
    x_next = a * np.sin(3 ** (y[-1] ** 2) - x[-1] ** 3)
    y_next = b * np.cos(3 ** (x[-1] ** 2) - y[-1] ** 3)
    x.append(x_next)
    y.append(y_next)

x_series = np.array(x)
y_series = np.array(y)

# ==== 2. 构造LSTM训练数据 ====
seq_len = 10
train_len = 500
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def prepare_data(series):
    X, Y = [], []
    for i in range(train_len - seq_len):
        X.append(series[i:i+seq_len])
        Y.append(series[i+seq_len])
    X = torch.tensor(X, dtype=torch.float32).unsqueeze(-1).to(device)
    Y = torch.tensor(Y, dtype=torch.float32).unsqueeze(-1).to(device)
    return X, Y

Xx, Yx = prepare_data(x_series)
Xy, Yy = prepare_data(y_series)

# ==== 3. 定义LSTM模型 ====
class BiLSTMForecast(nn.Module):
    def __init__(self, hidden_size):
        super(BiLSTMForecast, self).__init__()
        self.hidden_size = hidden_size
        
        # 修改为双向LSTM (bidirectional=True)
        self.lstm = nn.LSTM(input_size=1, 
                           hidden_size=hidden_size,
                           num_layers=2,
                           batch_first=True,
                           bidirectional=True)  # 关键修改
        
        # 全连接层输入维度需要x2 (因为双向LSTM会拼接正向和反向的hidden states)
        self.fc = nn.Linear(hidden_size * 2, 1)  # 修改输出维度
    
    def forward(self, x):
        # x shape: (batch_size, seq_len, input_size=1)
        
        # LSTM层输出
        lstm_out, _ = self.lstm(x)  # lstm_out shape: (batch_size, seq_len, hidden_size*2)
        
        # 只取最后一个时间步的输出
        last_out = lstm_out[:, -1, :]  # shape: (batch_size, hidden_size*2)
        
        # 全连接层
        output = self.fc(last_out)  # shape: (batch_size, 1)
        
        return output


# ==== 4. 模型训练函数（修改：添加学习率余弦下降和loss记录）====
def train_model(X_train, Y_train, model_name, hidden_size=64, epochs=500, lr=1e-2):
    model = BiLSTMForecast(hidden_size).to(device)
    optimizer = optim.Adam(model.parameters(), lr=lr)
    criterion = nn.MSELoss()
    
    # 新增：创建学习率调度器 - 余弦退火
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-4)
    
    # 新增：创建loss记录列表
    loss_history = []
    lr_history = []  # 新增：记录学习率变化
    
    for ep in range(epochs):
        model.train()
        optimizer.zero_grad()
        output = model(X_train)
        loss = criterion(output, Y_train)
        loss.backward()
        optimizer.step()
        
        # 新增：更新学习率
        scheduler.step()
        
        # 新增：记录每个epoch的loss和学习率
        loss_history.append(loss.item())
        current_lr = optimizer.param_groups[0]['lr']
        lr_history.append(current_lr)
        
        # 每50个epoch打印一次进度（新增学习率显示）
        if (ep+1) % 50 == 0:
            print(f'{model_name} - Epoch [{ep+1}/{epochs}], Loss: {loss.item():.6f}, LR: {current_lr:.2e}')
    
    # 新增：将loss和学习率保存到CSV
    df = pd.DataFrame({
        'epoch': range(1, epochs+1),
        'loss': loss_history,
        'learning_rate': lr_history
    })
    df.to_csv(f'{model_name}_training_history.csv', index=False)
    
    return model


# 修改：添加模型名称参数
model_x = train_model(Xx, Yx, 'model_x')
model_y = train_model(Xy, Yy, 'model_y')

# ==== 5. 使用训练好的模型预测后续序列 ====
def predict_future(model, series, pred_len):
    model.eval()
    window = list(series[train_len - seq_len:train_len])
    preds = []
    for _ in range(pred_len):
        inp = torch.tensor(window, dtype=torch.float32).unsqueeze(0).unsqueeze(-1).to(device)
        with torch.no_grad():
            pred = model(inp).item()
        preds.append(pred)
        window.pop(0)
        window.append(pred)
    return np.concatenate([series[:train_len], preds])

pred_len = 500
x_new = predict_future(model_x, x_series, pred_len)
y_new = predict_future(model_y, y_series, pred_len)

# ==== 6. 绘图，完全复现Fig.4风格 ====
series_pairs = [('x', x_series, x_new),
                ('y', y_series, y_new)]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
labels = ['(a)', '(b)']

for idx, (var, orig, pred) in enumerate(series_pairs):
    ax = axes[idx]
    ax.plot(orig, color='blue', label=f'${var}_n$')
    ax.plot(pred, color='red', linestyle='--', label=f'${var}_n\prime$')
    ax.axvline(train_len, color='gray', linestyle=':', linewidth=1)
    ax.set_xlim(0, len(orig))
    ax.set_xlabel('$n$')
    ax.set_ylabel(f'${var}$')
    ax.legend(loc='upper right')
    ax.text(0.05, 0.9, labels[idx], transform=ax.transAxes, fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from scipy import spatial

# ================= 1.生成混沌序列（Lorenz系统）=================
def lorenz(t, state, sigma=10, beta=8/3, rho=28):
    x, y, z = state
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return [dx, dy, dz]

# 数值积分求解
t_span = (0, 50)
t_eval = np.linspace(*t_span, 5000)
sol = solve_ivp(lorenz, t_span, [1,1,1], t_eval=t_eval)
original_series = sol.y[0]  # 取x分量作为原始序列

# ================= 2.数据预处理 =================
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(original_series.reshape(-1,1))

# 创建监督学习数据集
def create_dataset(data, look_back=20):
    X, Y = [], []
    for i in range(len(data)-look_back-1):
        X.append(data[i:(i+look_back), 0])
        Y.append(data[i + look_back, 0])
    return np.array(X), np.array(Y)

X, y = create_dataset(scaled_data)
X = X.reshape(X.shape[0], X.shape[1], 1)  # 调整为LSTM输入格式

# ================= 3.构建LSTM模型 =================
model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(LSTM(32))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# 训练模型
history = model.fit(X, y, epochs=50, batch_size=64, validation_split=0.2)

# ================= 4.递归预测序列 =================
def recursive_predict(model, seed, steps=1000):
    predictions = []
    current_seq = seed.copy()
    for _ in range(steps):
        pred = model.predict(current_seq.reshape(1, len(seed), 1))[0,0]
        predictions.append(pred)
        current_seq = np.roll(current_seq, -1)
        current_seq[-1] = pred
    return scaler.inverse_transform(np.array(predictions).reshape(-1,1)).flatten()

# 使用最后20个点作为初始种子
predicted_series = recursive_predict(model, scaled_data[-20:])

# ================= 5.计算李雅普诺夫指数 =================
def compute_lyapunov(series, emb_dim=3, tau=5, nn=10):
    """使用Rosenstein算法计算最大李雅普诺夫指数"""
    n = len(series) - (emb_dim-1)*tau
    emb = np.array([series[i:i+(emb_dim-1)*tau+1:tau] for i in range(n)])
    tree = spatial.cKDTree(emb)
    
    dists = []
    for i in range(0, len(emb), 10):  # 每隔10个点采样
        dist, idx = tree.query(emb[i], k=nn+1)
        dists.append(np.mean(dist[1:]))  # 排除自身
    
    d = np.log(np.abs(np.array(dists)))
    x = np.arange(len(d))
    return np.polyfit(x, d, 1)[0]

# 计算指数
le_original = compute_lyapunov(original_series)
le_predicted = compute_lyapunov(predicted_series)

print(f"Original Lyapunov exponent: {le_original:.4f}")
print(f"Predicted Lyapunov exponent: {le_predicted:.4f}")

# ================= 6.绘制李雅普诺夫图 =================
plt.figure(figsize=(14,6))

# 原序列相空间
plt.subplot(121, projection='3d')
plt.plot(original_series[:-2], original_series[1:-1], original_series[2:], lw=0.4)
plt.title(f'Original System (λ={le_original:.3f})')

# 预测序列相空间
plt.subplot(122, projection='3d')
plt.plot(predicted_series[:-2], predicted_series[1:-1], predicted_series[2:], lw=0.4)
plt.title(f'Predicted System (λ={le_predicted:.3f})')

plt.tight_layout()
plt.savefig('lyapunov_comparison.png', dpi=300)
plt.show()
```
